## Data Preprocessing

```python
data.head() # shownuing the top 5 data
```

```python
data.isnull().sum() # check the null value
```

* Here we can see the total 2 null values columns is Route or Total_Stops.

```python
data.Route.mode() # find the median
```

```python
# Handle the Null Value
data.loc[data["Route"].isnull(),"Route"] = "DEL → BOM → COK"
```

```python
data.Total_Stops.mode() # find the median
```

```python
# Handle the Null Value
data.loc[data["Total_Stops"].isnull(),"Total_Stops"] = "1 stop"
```

```python
data.isnull().sum() # check the null value
```

```python
data.info() # check the dataset information
```

```python
data["Date_of_Journey"] = data["Date_of_Journey"].astype("datetime64[ns]") # change the datatype
```

```python
data["Dep_Time"] = pd.to_datetime(data["Dep_Time"]) # change the datatype
data["Arrival_Time"] = pd.to_datetime(data["Arrival_Time"]) # change the datatype
```

```python
data['Duration'] = data['Duration'].str.replace('h', ' hours ').str.replace('m', ' minutes')
data['Duration'] = pd.to_timedelta(data['Duration'])
```

```python
data.info() # check the dataset information
```

```python
data.head() # shownig the top 5 data
```

```python
data["Day"] = pd.to_datetime(data["Date_of_Journey"]).dt.day # create the day column
data["Month"] = pd.to_datetime(data["Date_of_Journey"]).dt.month # create the month column
data["Year"] = pd.to_datetime(data["Date_of_Journey"]).dt.year # create the year column
```

```python
data.head() # shownig the top 5 data
```

```python
data["Dep_Hour"] = data["Dep_Time"].dt.hour # craete a Hour Column
data["Dep_Min"] = data["Dep_Time"].dt.minute # craete a Mintue Column
```

```python
data["Arri_Hour"] = data["Arrival_Time"].dt.hour # craete a Hour Column
data["Arri_Min"] = data["Arrival_Time"].dt.minute # craete a Mintue Column
```

```python
data["Duration_Hour"] = data["Duration"].dt.components['hours'] # craete a Hour Column
data["Duration_Min"] = data["Duration"].dt.components['minutes'] # craete a Mintue Column
```

```python
data.head() # shownig the top 5 data
```

```python
from sklearn.preprocessing import OneHotEncoder  # import the Onehot encoding
OE = OneHotEncoder()  # store the label encoding
df = pd.get_dummies(data["Airline"],dtype="int",prefix="Airlines") # tranform the data
df1 = pd.get_dummies(data["Source"],dtype="int",prefix="Source")  # tranform the data
df2 = pd.get_dummies(data["Destination"],dtype="int",prefix="Destination")  # tranform the data
```

```python
data1 = pd.concat([data,df,df1,df2],axis=1) # concate the four dataset
```

```python
# show the all column
pd.set_option("display.max_column",None)
```

```python
data1.head() # showning the top 5 data
```

```python
data1.head() # showning the top 5 value
```

```python
data1.Total_Stops.unique() # check the total unique value
```

```python
data1.loc[data1["Total_Stops"] == "non-stop","Total_Stops"] = 0  # change the total stop value
data1.loc[data1["Total_Stops"] == "1 stop","Total_Stops"] = 1  # change the total stop value
data1.loc[data1["Total_Stops"] == "2 stops","Total_Stops"] = 2  # change the total stop value
data1.loc[data1["Total_Stops"] == "3 stops","Total_Stops"] = 3  # change the total stop value
data1.loc[data1["Total_Stops"] == "4 stops","Total_Stops"] = 4  # change the total stop value
```

```python
data1["Total_Stops"] = data1["Total_Stops"].astype(int) # change the dataype 
```

```python
data1.head()  # showning the top 5 data
```

```python
data1.Year.unique()  # check the unique value
```

```python
# drop the categorical column 
data2 = data1.drop(["Airline","Source","Destination","Date_of_Journey","Route","Dep_Time","Arrival_Time","Duration","Additional_Info","Year"],axis=1)
data2
```

## Gradiant Boosting Algorithm

```python
from sklearn.ensemble import GradientBoostingRegressor # import the Gradinat boosting Algorithm
GB = GradientBoostingRegressor() # call the Gradinat boosting algorithm
GB.fit(x_train,y_train) # train the data
```

```python
y_gb = GB.predict(x_test) # predict the tesning data
```

```python
r2_score(y_test,y_gb) # check the model performance
```

```python
model_gb = GradientBoostingRegressor() #call the model
# Hyperparamter of GradientBOssting
gb_grid = {
    'n_estimators': [100, 200, 300],          # Number of boosting stages
    'learning_rate': [0.01, 0.1, 0.2],        # Step size shrinkage
    'max_depth': [3, 5, 7],                   # Maximum depth of the individual regression estimators
    'min_samples_split': [2, 5, 10],          # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],            # Minimum number of samples at a leaf node
    'max_features': ['sqrt', 'log2', None],   # Number of features to consider when looking for the best split
    'subsample': [0.8, 1.0],                  # Fraction of samples used for fitting individual base learners
    'loss': ['squared_error', 'absolute_error']
}
gb_search = GridSearchCV(estimator=model_gb,param_grid=gb_grid,cv=3,scoring="r2",verbose=2,n_jobs=-1) # Create the Gridsearch cv tuning
gb_search.fit(x_train,y_train) # train the model
```
